```python
# Readability Reflection Themes Bar Chart

import numpy as np
import matplotlib.pyplot as plt

# --- Data from your LaTeX table ---
themes = [
    "Concatenation",
    "Interpolation",
    "Context Dependent",
    "Both"
]
counts = [94, 63, 7, 6]
colors = ['#1f6f6f', '#54a1a1', "#9fd4d4", '#cce8e8']
# --- Sort by count (descending) for clearer presentation ---
pairs = sorted(zip(counts, themes), reverse=True)
counts_sorted, themes_sorted = zip(*pairs)

# --- Build a horizontal bar chart (better for long labels) ---
fig, ax = plt.subplots(figsize=(10, 4))
bars = ax.barh(
    range(len(themes_sorted)),
    counts_sorted,
    edgecolor="black",   # black edges
    color=colors,         # gray fill
    linewidth=1.0
)

# Axis labels (18 pt)
ax.set_xlabel("Number of Responses", fontsize=18)
ax.set_ylabel("Theme", fontsize=18)

# Y ticks (18 pt)
ax.set_yticks(np.arange(len(themes_sorted)))
ax.set_yticklabels(themes_sorted)
ax.tick_params(axis="both", which="major", labelsize=18)

# Largest bar on top
ax.invert_yaxis()

# --- Annotate each bar with its value (18 pt) ---
for bar, val in zip(bars, counts_sorted):
    ax.text(
        bar.get_width() + max(counts_sorted) * 0.01,
        bar.get_y() + bar.get_height() / 2,
        f"{val}",
        va="center",
        ha="left",
        fontsize=18
    )

# --- Clean visual style for publication ---
for spine in ["top", "right"]:
    ax.spines[spine].set_visible(False)
ax.xaxis.grid(True, linestyle=":", linewidth=0.8)
ax.set_axisbelow(True)

plt.tight_layout()

# --- Save high-resolution output ---
png_path = "analysis/figures/Fig_readability_reflection.png"
pdf_path = "analysis/figures/Fig_readability_reflection.pdf"
plt.savefig(png_path, dpi=1200, bbox_inches="tight")
plt.savefig(pdf_path, dpi=1200, bbox_inches="tight")
plt.show()
```
